```python
#in this we used a classifier xml file from open cv git which helps us in feature classification
face_cascade = cv2.CascadeClassifier('.\open cv\haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('.\open cv\haarcascade_eye.xml')

faces = face_cascade.detectMultiScale(img, 1.3, 5)
faces
```

```python
#this helps us to store the data about the face which are pixels of the face 
# and at which space the face is there in the picture 
# hence they store the coordinates of the face
(x, y, w, h) = faces[0]
x, y, w, h
```

```python
#this function is to draw a rectangle around the face detected for better 
#classification by giving input of the coordinates generated abov
face_img = cv2.rectangle(img,(x,y),(x+w, y+h),(255, 0, 0), 2)
plt.imshow(face_img)
```

```python
#here is the for loop for  deteccting faces in an image containing multiple 
#face along with the eyes as well and also making rectangle around the face 
#and eyes detected 
cv2.destroyAllWindows()
for (x, y, w, h) in faces:
    face_img = cv2.rectangle(img,(x,y),(x+w, y+h),(255, 0, 0), 2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for(ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew, ey+eh),(0, 255, 0), 2)

plt.figure()
plt.imshow(face_img, cmap='gray')
plt.show()
```

```python
#this is to plot image inside the rectangle only as it is only required for face
#detection
plt.imshow(roi_color, cmap='gray')
```

```python
#this is function to summarize all the things we did above to any image we 
#want by just giving the image path
def crop(image_path):
    img  = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes)>=2:
            return roi_color
    
```

# Now the data cleaning is done we move on to feature extraction

from here we will do wavelet transform which is the most effective way of extracting literatures

```python
#fourier transform 
```

```python
import numpy as np 
import pywt
import cv2

def wavelet(img, mode='haar', level =1):
    imArray= img 
    #datattype conversion
    #convert to gray scale 
    imArray = cv2.cvtColor(imArray, cv2.COLOR_RGB2GRAY)
    imArray= np.float32(imArray)
    imArray/=255
    #compute coeffecients
    coeffs = pywt.wavedec2(imArray, mode, level = level)
    #process coeffecients
    coeffs_H= list(coeffs)
    coeffs_H[0]*= 0
    
    #reconstruction
    imArray_H = pywt.waverec2(coeffs_H, mode)
    imArray_H *=255
    imArray_H= np.uint8(imArray_H)
    return imArray_H
    
    
```

```python
#assigning each celeb a value and we access the celeb using dictionary 
class_dict ={}
count=0
for celeb_name in celeb_file_name_dict.keys():
    class_dict[celeb_name]= count
    count = count+1
    
class_dict
```

```python
X=[]
y=[]
for celeb_name, cropped_files in celeb_file_name_dict.items():#now we go through all the cropped images
    for images in cropped_files:
        img = cv2.imread(images)
        if img is None: #as we have deleted some images ourselved for better data hence we add this if statement 
            continue    #if the image is not found on the path it does not show any error
        scaled_img = cv2.resize(img, (32, 32))
        im_har = wavelet(img, 'db1', 5)
        scaled_img_har = cv2.resize(im_har,(32,32)) #resize all the images so that the model can run smoothly
        combined_img = np.vstack((scaled_img.reshape(32*32*3, 1), scaled_img_har.reshape(32*32, 1))) #here we stack the cropped and wavelet tranformed image together 
        X.append(combined_img)
        y.append(class_dict[celeb_name]) 
```

```python
X= np.array(X).reshape(len(X), 4096).astype(float)
X.shape
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state =0)
pipe = Pipeline([('scaler', StandardScaler()), ('svc', SVC(kernel ='rbf', C=10))])
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

```python
print(classification_report(y_test, pipe.predict(X_test)))
```

## Now we are going to apply grid search CV it is basically used to hypertune parameters and get the best model

```python
# for example like above we applied SVM now we change the parameters in SVM like
#kernel and c value to know which one is best for our data 
#cross-validation tutorial
#validation set and test set 
```

```python
from sklearn import svm 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

```python
#so here what we do is make a dictionary which contains different models and 
# inside those models we store the parameters that we want to train our model on
# so basically what it does is run different models along with different parameters
# of that model and find the best model along with the most suitable parameters
model_params ={
    'svm':{
        'model':svm.SVC(gamma='auto', probability= True),
        'params':{
            'svc__C':[1,10,100,1000], 
            'svc__kernel':['rbf', 'linear']
        }
    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'randomforestclassifier__n_estimators': [1,5,10]
        }
    },
     'logistic_regression':{
        'model':LogisticRegression(solver='liblinear', multi_class='auto'),
        'params':{
            'logisticregression__C': [1,5,10]
        }
    },
    
}
```

Validation datasets contain different samples to evaluate trained ML models. It is still possible to tune and control the model at this stage. Working on validation data is used to assess the model performance and fine-tune the parameters of the model. This becomes an iterative process wherein the model learns from the training data and is then validated and fine-tuned on the validation set. A validation dataset tells us how well the model is learning and adapting, allowing for adjustments and optimizations to be made to the model's parameters or hyperparameters before it's finally put to the test.

```python
#here we iterate through the dictionary we made above 
#in the loop the grid search function find score of the model by running the 
#train dataset along with the validation dataset 

scores =[]
best_estimators={}
import pandas as pd
for algo, mp in model_params.items():
    pipe = make_pipeline(StandardScaler(), mp['model'])
    clf = GridSearchCV(pipe, mp['params'], cv=5, return_train_score=False)
    clf.fit(X_train, y_train)
    scores.append({
        'model':algo,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    best_estimators[algo]= clf.best_estimator_
    
df = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df
```
